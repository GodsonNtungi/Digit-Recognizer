```python
import tensorflow as tf
import tensorflow.keras.datasets as tfd
from tensorflow import keras
from keras import Sequential, layers
```

```python
data = tfd.mnist.load_data(path='mnist.npz')
```

```python
## (data is a tuple (x_train,y_train),(x_test,y_test))
train = data[0]
test = data[1]
```

```python
X = train[0]
y = train[1]
x_test = test[0]
y_test = test[1]
```

```python
rescale = Sequential([
    layers.Rescaling(1./255,input_shape=(28,28,1))
])
```

```python
augumentation = Sequential([
    layers.RandomRotation(0.4),
])
```

```python
from keras.layers.pooling import MaxPool2D
model = Sequential([
    rescale,
    #augumentation,
    layers.Conv2D(16,3,activation='relu'),
    layers.MaxPool2D(),
    layers.Conv2D(32,3,activation='relu'),
    layers.MaxPool2D(),
    layers.Dropout(0.3),
    layers.Flatten(),
    layers.Dense(128,activation='relu'),
    layers.BatchNormalization(),
    layers.Dropout(0.3),
    layers.Dense(128,activation='relu'),
    layers.BatchNormalization(),
    layers.Dropout(0.3),
    layers.Dense(10,activation='softmax')
])
```

```python
model.compile(optimizer='adam',
              loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
              metrics=['accuracy'])
model.build()
model.summary()
```

```python
earlystopping = tf.keras.callbacks.EarlyStopping(
    monitor='val_accuracy', min_delta=0.01, patience=50, verbose=0,
    mode='max'
)
```

```python
history = model.fit(X,y,validation_split=0.2,epochs=100,callbacks=earlystopping)
```

```python
model.fit(X,y,validation_split=0.2,epochs=200,initial_epoch=100, callbacks=earlystopping)
```

```python
model.evaluate(x_test,y_test)
```

```python
model.save('digitrecognizer99.h5')
```
